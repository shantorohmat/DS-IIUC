```python
def meta_data_scrap(driver):
    try:
        meta_data = driver.find_element(by=By.XPATH, value='//*[@id="app-wrapper"]/div[1]/div[2]/div[2]/div[3]/div[2]/div/div[1]/div/div[2]/div[2]')

        md = meta_data.find_elements(by=By.TAG_NAME,value='div')
        meta_dic = {}
        for m in md:
            m = m.text
            if '\n' in m:
                key, val = m.split('\n')[0], m.split('\n')[1]
                meta_dic[key] = val
        print("Meta Data: ", meta_dic)
        return meta_dic
    except:
        print('No meta data found')        
        return 'N/A'
```

```python
def title_scrp(driver):
    try:
        title = driver.find_element(By.XPATH,'//*[@id="app-wrapper"]/div[1]/div[2]/div[2]/div[3]/div[1]/div[1]/div/h1').text
        print("Title: ", title)
        return title
    except:
        print("Product Not found. May be sold out!")
        return None
```

```python
def time_location_scrap(driver):
    try:
        time_location =  driver.find_element(By.XPATH,'//*[@id="app-wrapper"]/div[1]/div[2]/div[2]/div[3]/div[1]/div[1]/div/span').text
        print("Time & Location: ", time_location)
        return time_location
    except:
        print("nothing")
        return None
        
```

```python
def price_scrp(driver):
    try:
        price = driver.find_element(By.XPATH,'//*[@id="app-wrapper"]/div[1]/div[2]/div[2]/div[3]/div[2]/div/div[1]/div/div[2]/div[1]/div/div[1]').text
        print("Price: ", price)
        return price
    except:
        print("NOthing")
        return None
```

```python
def feature_scrp(driver):
    try:
        feature = driver.find_element(By.XPATH,'//*[@id="app-wrapper"]/div[1]/div[2]/div[2]/div[1]/div[1]/div/div[1]/div/div[2]/div[3]').text
        print('Features: ', feature)
        return feature                   
    except:
        return None
```

```python
def description_scrp(driver):
    try:
        description = driver.find_element(By.XPATH,'//*[@id="collapsible-content-0"]/ul/div').text
        print("Description: ", description)
        return description
    except:
        return None
```

```python
def show_more_click(driver):    
    try:
        driver.find_element(by=By.CLASS_NAME, value='show-text--13FcL').click()
#         s_driver.find_element(by=By.XPATH, value='//*[@id="app-wrapper"]/div[1]/div[2]/div[2]/div[3]/div[2]/div/div[1]/div/div[2]/div[3]/div/div[2]/div/div[2]/button').click()
        time.sleep(2)
    except:
        print("No 'Show more' button")
```

```python
def open_driver():
    op = webdriver.ChromeOptions()
    op.add_argument('headless')
    s = Service('chromedriver')
    driver = webdriver.Chrome(service=s, options=op)
    return driver
```

```python
def image_scrp(driver, category):
    try:
        images = driver.find_element(by=By.XPATH, value='//*[@id="app-wrapper"]/div[1]/div[2]/div[2]/div[3]/div[2]/div/div[1]/div/div[1]/div/div/ul')
        img_path = []
        for img in images.find_elements(by=By.TAG_NAME,value='img'):
            img_urls = img.get_attribute('src') ## main link = img_urls
## main link=https://i.bikroy-st.com/xiaomi-poco-c31-intck-offer-new-for-sale-rangpur/d76b227e-d700-4a54-8ac0-61d9bc0c9e3b/100/70/cropped.jp
                                                                ##[start:end:step] 
                                                                ## -3,-2,-1 index er charater gula bad dia new string.
            img_urls_firstname = '/'.join(img_urls.split('/')[:-3])## main link thak 3 ta character bad
            
## firstname=https://i.bikroy-st.com/xiaomi-poco-c31-intck-offer-new-for-sale-rangpur/d76b227e-d700-4a54-8ac0-61d9bc0c9e3b
            
            img_urls_rename = img_urls_firstname + '/620/466/fitted.jpg'## main link e 3 ta character plus
                                                                        ## ei khane ekta new link create korlo
# img er jonno main link jar thaka amr pic scrap korbo
#https://i.bikroy-st.com/xiaomi-poco-c31-intck-offer-new-for-sale-rangpur/d76b227e-d700-4a54-8ac0-61d9bc0c9e3b/620/466/fitted.jpg

            img_name = '@'.join(img_urls_firstname.split('/')[-2:])## -2 index thaka string ta start hoise. and itz img_name
# firstname er string er -2,-1 index er character nia img_name
#img_name = xiaomi-poco-c31-intck-offer-new-for-sale-rangpur@d76b227e-d700-4a54-8ac0-61d9bc0c9e3b
            r = requests.get(img_urls_rename)
            i = Image.open(BytesIO(r.content))
            i.save(f"E:/DS - IIUC/Python/classes/second class scraping/bikroy_dotcom/image/{category}/{img_name}.jpg")
            img_path.append(img_name)
        return img_path
    except:
        print('No images')        
        return 'N/A'
```

```python
def product_scrap(driver, url, category):
    driver.get(url)
    time.sleep(5)

    # single page data scrapping
    title = title_scrp(driver)
    # if title not found (product sold) don't proceed further for other data
    if not title:
        return [None]*7
    # scrapping time and location
    time_location = time_location_scrap(driver)
    
    # scrapping price
    price = price_scrp(driver)
    
    # scrapping all images
    images = image_scrp(driver, category)
    

    # extractig meta data
    meta_data = meta_data_scrap(driver)
        
    # click on show_more button
    show_more_click(driver)

    # extracting feature data
    feature = feature_scrp(driver)
    
    # extracting description
    description = description_scrp(driver)
    
    return title, time_location, price, images, meta_data, feature, description
```

```python
category = 'mobile'
with open('E:/DS - IIUC/Python/classes/second class scraping/bikroy_dotcom/url.txt', 'r') as f:
    single_url = f.readlines()
    titles, locations, prices, img_lst, meta_lst,  feature_lst, desc_lst = [], [], [], [], [], [], []
    cnt = 1
#     print(single_url)
    for s in single_url:
        print(cnt, s)
        driver = open_driver()
        
        title, time_location, price, images, meta_data, feature, description = product_scrap(driver, s, category)
        if title:
            titles.append(title)
            locations.append(time_location)
            prices.append(price)
            img_lst.append(images)
            meta_lst.append(meta_data)
            feature_lst.append(feature)
            desc_lst.append(description) 

            cnt += 1
            print("=============================================")
#             time.sleep(10)
#             clear_output(wait=True)
```
